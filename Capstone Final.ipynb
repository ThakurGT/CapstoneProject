###  Here we extract data from _FourSquare API_

```python
CLIENT_ID='dffrererewedwdwdww'
CLIENT_SECRET='wwwwwwwwww'
VERSION='20190726'
radius='600'
url='https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}'.format(CLIENT_ID,CLIENT_SECRET,VERSION,latitude,longitude,radius)

data=requests.get(url).json()
data # data in the form of json
```

```python
nb_data=data['response']['groups'][0]['items']
nb_data[0]
```

### Here we normalize data in the form of Data Frames and clean it

```python
df=json_normalize(nb_data) # Normalise data to dataframe

cols=['venue.name','venue.categories']+[cl for cl in df.columns if cl.startswith('venue.location')]+['venue.id']# keep only required columns

df_filtered=df.loc[:,cols]
def get_category_type(row):
    try:
        categories_list=row['categories']
    except:
        categories_list=row['venue.categories']
    if len(categories_list)==0:
        return None
    else:
        return categories_list[0]['name']
    
df_filtered['categories']=df_filtered.apply(get_category_type,axis=1)

df_filtered.columns=[column.split('.')[-1] for column in df_filtered.columns]

print(df_filtered.shape)
df_filtered
```

#### We can see above are the listed interesting places nearby San Diego. This will help know where are the more attractive areas

### Now we will find the  restaurants(any) in the area and if there is any Indian Restaurant or not. Same way as above we will extract the data from Foursquare API

```python
search_query='Restaurant'
url2='https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&query={}'.format(CLIENT_ID,CLIENT_SECRET,VERSION,latitude,longitude,radius,search_query)

data2=requests.get(url2).json()
data2
```

```python
nb_data2=data2['response']['venues']
nb_data2[0]
```

### Following same procedusre as previous one we will noramlize the data to data frames and clean as per requirements

```python
df2=json_normalize(nb_data2)

cols=['name','categories']+[cl for cl in df2.columns if cl.startswith('location')]+['id']

df_filtered2=df2.loc[:,cols]
def get_category_type(row):
    try:
        categories_list=row['categories']
    except:
        categories_list=row['venues.categories']
    if len(categories_list)==0:
        return None
    else:
        return categories_list[0]['name']
    
df_filtered2['categories']=df_filtered2.apply(get_category_type,axis=1)

df_filtered2.columns=[column.split('.')[-1] for column in df_filtered2.columns]
print(df_filtered2.shape)
df_filtered2
```

### Map 1: Visualizing map for number of interesting places nearby

```python
venue_map=folium.Map(location=[latitude,longitude],zoom_start=16)

folium.Marker(
    [latitude,longitude],
    popup=add,
    icon=folium.Icon(color='red')
    ).add_to(venue_map)

for lat,lng,label in zip(df_filtered.lat, df_filtered.lng, df_filtered.name):
    label = folium.Popup(label, parse_html=True)
    folium.Marker(
    [lat,lng],
    popup=label,
    icon=folium.Icon(color='blue')
    ).add_to(venue_map)
    
venue_map
```

### Map 2: Vizualizing the density of restaurants nearby according to the defined radius

```python
venue_map2=folium.Map(location=[latitude,longitude],zoom_start=16)

folium.Marker(
    [latitude,longitude],
    popup=add,
    icon=folium.Icon(color='red')
    ).add_to(venue_map2)

for lat,lng,label in zip(df_filtered2.lat, df_filtered2.lng, df_filtered2.name):
    label = folium.Popup(label, parse_html=True)
    folium.Marker(
    [lat,lng],
    popup=label,
    icon=folium.Icon(color='blue')
    ).add_to(venue_map2)
    
venue_map2
```
